## Q2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
df_gold = refined_df[refined_df.Medal == 'Gold']
```

```python
df_gold
```

```python
gold_medal_list = []

for Sport in df_gold.Sport.unique():
    
    gold_medal_list.append([Sport, len(df_gold[df_gold.Sport == Sport])])

gold_medal_list = pd.DataFrame(gold_medal_list, columns =['Sport','Gold Medals'])

gold_medal_list = gold_medal_list.sort_values(by = 'Gold Medals', ascending = False).head()

gold_medal_list.plot(x = 'Sport', y = 'Gold Medals', kind = 'bar',color = 'gold', figsize = (8,6))

print ("The Sport having most number of gold medals is : ", gold_medal_list.Sport[0])
```

```python
gold_medal_list
```
